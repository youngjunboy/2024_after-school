```python
def add_self(c):
    if c==1:
        return 1
    else:
        return c + add_self(c-1)
    
res = add_self(17)
print(res)
```

```python
x = [1,2,3]
y = [1,3,2]
n = len(x)

sum_x=sum_y=0
for i in range(n):
    sum_x += x[i]
    sum_y += y[i]
    
mean_x = sum_x/n
mean_y = sum_y/n
mean_x, mean_y
sum_xx = sum_xy = 0

for i in range(n):
    sum_xx += (x[i]-mean_x)**2
    sum_xy += (x[i]-mean_x)*(y[i]-mean_y)
    
    beta = sum_xy/sum_xx
    alpha = mean_y-beta*mean_x
    alpha, beta
    
from matplotlib import pyplot as plt

reg_x = [0,1,2,3,4,5]
reg_y = [(alpha+beta*x) for x in reg_x]

plt.figure(figsize=(5,3))
plt.grid()
plt.scatter(x,y)
plt.plot(reg_x,reg_y,'r-')
plt.show()
```
