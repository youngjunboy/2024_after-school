```python
class Sort : 
    def swap(self,a,b):
        t = b
        b = a
        a = t
        return a,b
    
    def bubble_sort(self,arr):
        for i in range(len(arr)):
            for j in range(len(arr) - i - 1):
                if arr[j] > arr[j+1]:
                    arr[j], arr[j+1] = self.swap(arr[j], arr[j+1])
        return arr
    
arr = [4,5,2,6,1,3,7,8,9]
sort = Sort()
new_arr = sort.bubble_sort(arr)

print(new_arr)
```
