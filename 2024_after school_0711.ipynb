```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    '이름' : ['철수','영희','민수','지민','수지'],
    '수학' : [90,70,80,100,60],
    '영어' : [85,80,75,90,65],
    '과학' : [95,75,85,95,70],
}
df = pd.DataFrame(data)
# 과목별 평균 점수 계산
subject_means = df.mean(numeric_only=True)
print('과목별 평균 점수 : ')
print(subject_means)
# 학생별 평균 점수 계산
df['평균'] = df.mean(axis=1, numeric_only=True)
print('|n 학생별 평균 점수 : ')
print(df[['이름','평균']])
# 과목별 최고 점수와 최저 점수를 가진 학생 찾기
max_scores = df[['이름','수학','영어','과학']].set_index('이름').idxmax()
min_scores = df[['이름','수학','영어','과학']].set_index('이름').idxmin()
print('|n 과목별 최고 점수 학생 : ')
print(max_scores)
print('|n 과목별 최저 점수 학생 : ')
print(min_scores)
# 한글 폰트 설정
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
# 시각화 - 과목별 평균 점수 막대 그래프
subject_means.plot(kind='bar')
plt.title('과목별 평균 점수')
plt.xlabel('과목')
plt.ylabel('평균 점수')
plt.show()
```
